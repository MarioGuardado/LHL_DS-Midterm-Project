```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
#use data scraping to extract the largest 500 companies in the US, by market capitalization, from the website companiesmarketcap.com

#initialize a list for each column of the table
company_names = []
company_codes = []
market_caps = []
prices = []
today_changes = []
prices_30_days = []
countries = []

#the website displays at most 100 companies per page, then data must be scraped from the first five pages
for page in range(0,5):
    url = f"https://companiesmarketcap.com/usa/largest-companies-in-the-usa-by-market-cap/?page={page + 1}"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        table = soup.find('table', class_='marketcap-table')
        if table:
            #find all the 'td' elements that contain data, corresponging to table rows
            data_cells = table.find_all('td')            

            for i in range(0, len(data_cells), 7):
                # Extract information from each set of 7 data cells (each company's data)
                company = BeautifulSoup(str(data_cells[i + 1]), 'html.parser')
                company_name = company.find('div', class_='company-name').text
                company_code = company.find('div', class_='company-code').text

                company_market_cap = data_cells[i + 2].text.strip()
                company_price = data_cells[i + 3].text.strip()
                today_change = data_cells[i + 4].text.strip()
                price_30_days = data_cells[i + 5].text.strip()
                company_country = data_cells[i + 6].text.strip()

                # Append extracted data to lists
                company_names.append(company_name)
                company_codes.append(company_code)
                market_caps.append(company_market_cap)
                prices.append(company_price)
                today_changes.append(today_change)
                prices_30_days.append(price_30_days)
                countries.append(company_country)
                        
    else:
        print("Failed to fetch the web page.")
        
# Create a DataFrame to store all the companies
uscompanies = {
             "Company Name": company_names,
             "Company Code": company_codes,
             "Market Cap": market_caps,
             "Price": prices,
             "Today's Change": today_changes,
             "Price (30 days)": prices_30_days,
             "Country": countries
            }

top500_companiesUS = pd.DataFrame(uscompanies)

#save the DataFrame to a csv file in the data/companies list folder
top500_companiesUS.to_csv(f"{os.path.abspath('../data/companies list/top500MarketCap_US.csv')}")
```
